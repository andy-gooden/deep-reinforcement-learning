### Part 1: TD Control: Sarsa

In this section, you will write your own implementation of the Sarsa control algorithm.

Your algorithm has four arguments:
- `env`: This is an instance of an OpenAI Gym environment.
- `num_episodes`: This is the number of episodes that are generated through agent-environment interaction.
- `alpha`: This is the step-size parameter for the update step.
- `gamma`: This is the discount rate.  It must be a value between 0 and 1, inclusive (default value: `1`).

The algorithm returns as output:
- `Q`: This is a dictionary (of one-dimensional arrays) where `Q[s][a]` is the estimated action value corresponding to state `s` and action `a`.

Please complete the function in the code cell below.

(_Feel free to define additional functions to help you to organize your code._)

````python
def sarsa(env, num_episodes, alpha, gamma=1.0, plot_freq=20):
    Q = defaultdict(lambda: np.zeros(env.nA))  # init action-value fn (empty dict of arrays)
    
    tmp_scores = deque(maxlen=plot_freq)      # deque for keeping track of scores
    avg_scores = deque(maxlen=num_episodes)   # average scores over every plot_every episodes
    
    for i_episode in range(1, num_episodes+1): # loop over episodes
        score = 0
        epsilon = max(1.0 / i_episode, .0001)   # for epsilon greedy policy
        done = False
        
        state = env.reset()
        action = epsilon_greedy(Q, state, env.action_space, epsilon)
        
        while not done:
            next_state, reward, done, info  = env.step(action)
            score += reward
            
            if reward == -100:  # cliff state, but done is broken
                done = True
            
            if not done:
                next_action = epsilon_greedy(Q, next_state, env.action_space, epsilon) # epsilon-greedy action
                Q[state][action] = Q[state][action] + alpha * (reward + gamma * Q[next_state][next_action] - Q[state][action])
                action = next_action
                state = next_state
            else:
                Q[state][action] = Q[state][action] + alpha * (reward - Q[state][action])
                tmp_scores.append(score)    # append score
                #print(f"S:{state} A:{action} R:{reward}, S:{next_state} A:{next_action} Done:{done}")

        if (i_episode % plot_freq == 0):
            avg_scores.append(np.mean(tmp_scores))
        
        if i_episode % 100 == 0:
            print("\rEpisode {}/{}".format(i_episode, num_episodes), end="")
            sys.stdout.flush()
            
    # plot performance
    plt.plot(np.linspace(0,num_episodes,len(avg_scores),endpoint=False), np.asarray(avg_scores))
    plt.xlabel('Episode Number')
    plt.ylim([-250,0])
    plt.ylabel('Average Reward (Over Next %d Episodes)' % plot_freq)
    plt.show()
    # print best 100-episode performance
    print(('Best Average Reward over %d Episodes: ' % plot_freq), np.max(avg_scores))
    
    return Q

def epsilon_greedy(Q, state, action_space, eps):
    """Selects epsilon-greedy action for supplied state.
    
    Params
    ======
        Q (dictionary): action-value function
        state (int): current state
        action_space (spaces obj): action space for environment
        eps (float): epsilon
    """
    if np.random.random() > eps: # select greedy action with probability epsilon
        return np.argmax(Q[state])
    else:                     # otherwise, select an action randomly
        return action_space.sample()
````

````python
# obtain the estimated optimal policy and corresponding action-value function
Q_sarsa = sarsa(env, 5000, .01)

# print the estimated optimal policy
policy_sarsa = np.array([np.argmax(Q_sarsa[key]) if key in Q_sarsa else -1 for key in np.arange(48)]).reshape(4,12)
check_test.run_check('td_control_check', policy_sarsa)
print("\nEstimated Optimal Policy (UP = 0, RIGHT = 1, DOWN = 2, LEFT = 3, N/A = -1):")
print(policy_sarsa)

# plot the estimated optimal state-value function
V_sarsa = ([np.max(Q_sarsa[key]) if key in Q_sarsa else 0 for key in np.arange(48)])
plot_values(V_sarsa)
````

````python
def q_learning(env, num_episodes, alpha, gamma=1.0, plot_freq=20, eps_min=.001):
    Q = defaultdict(lambda: np.zeros(env.nA))  # init action-value fn (empty dict of arrays)
    
    tmp_scores = deque(maxlen=plot_freq)      # deque for keeping track of scores
    avg_scores = deque(maxlen=num_episodes)   # average scores over every plot_every episodes
    
    for i_episode in range(1, num_episodes+1): # loop over episodes
        score = 0
        epsilon = max(1.0 / i_episode, eps_min)   # for epsilon greedy policy
        done = False
        
        state = env.reset()
        
        while not done:
            action = epsilon_greedy(Q, state, env.action_space, epsilon)
            next_state, reward, done, info  = env.step(action)
            score += reward
            
            if reward == -100:  # cliff state, but done is broken
                done = True
            
            if not done:
                Q[state][action] = Q[state][action] + alpha * (reward + gamma * np.max(Q[next_state]) - Q[state][action])
                state = next_state
            else:
                Q[state][action] = Q[state][action] + alpha * (reward - Q[state][action])
                tmp_scores.append(score)    # append score
                #print(f"S:{state} A:{action} R:{reward}, S:{next_state} A:{next_action} Done:{done}")

        if (i_episode % plot_freq == 0):
            avg_scores.append(np.mean(tmp_scores))
        
        if i_episode % 100 == 0:
            print("\rEpisode {}/{}".format(i_episode, num_episodes), end="")
            sys.stdout.flush()
            
    # plot performance
    plt.plot(np.linspace(0,num_episodes,len(avg_scores),endpoint=False), np.asarray(avg_scores))
    plt.xlabel('Episode Number')
    plt.ylim([-250,0])
    plt.ylabel('Average Reward (Over Next %d Episodes)' % plot_freq)
    plt.show()
    # print best 100-episode performance
    print(('Best Average Reward over %d Episodes: ' % plot_freq), np.max(avg_scores))
    
    return Q
````

### Part 3: TD Control: Expected Sarsa

In this section, you will write your own implementation of the Expected Sarsa control algorithm.

Your algorithm has four arguments:
- `env`: This is an instance of an OpenAI Gym environment.
- `num_episodes`: This is the number of episodes that are generated through agent-environment interaction.
- `alpha`: This is the step-size parameter for the update step.
- `gamma`: This is the discount rate.  It must be a value between 0 and 1, inclusive (default value: `1`).

The algorithm returns as output:
- `Q`: This is a dictionary (of one-dimensional arrays) where `Q[s][a]` is the estimated action value corresponding to state `s` and action `a`.

Please complete the function in the code cell below.

(_Feel free to define additional functions to help you to organize your code._)

````python
def expected_sarsa(env, num_episodes, alpha, gamma=1.0, plot_freq=20, eps_min=.001):
    Q = defaultdict(lambda: np.zeros(env.nA))  # init action-value fn (empty dict of arrays)
    
    tmp_scores = deque(maxlen=plot_freq)      # deque for keeping track of scores
    avg_scores = deque(maxlen=num_episodes)   # average scores over every plot_every episodes
    
    for i_episode in range(1, num_episodes+1): # loop over episodes
        score = 0
        epsilon = max(1.0 / i_episode, eps_min)   # for epsilon greedy policy
        done = False
        
        state = env.reset()
        
        while not done:
            action = epsilon_greedy(Q, state, env.action_space, epsilon)
            next_state, reward, done, info  = env.step(action)
            score += reward
            
            if reward == -100:  # cliff state, but done is broken
                done = True
            
            if not done:
                action_probs = np.ones(env.nA) * epsilon / env.nA
                action_probs[np.argmax(Q[state])] += (1 - epsilon)
                next_state_expected_value = np.dot(Q[next_state], action_probs)
                Q[state][action] = Q[state][action] + alpha * (reward + gamma * next_state_expected_value - Q[state][action])
                state = next_state
            else:
                Q[state][action] = Q[state][action] + alpha * (reward - Q[state][action])
                tmp_scores.append(score)    # append score
                #print(f"S:{state} A:{action} R:{reward}, S:{next_state} A:{next_action} Done:{done}")

        if (i_episode % plot_freq == 0):
            avg_scores.append(np.mean(tmp_scores))
        
        if i_episode % 100 == 0:
            print("\rEpisode {}/{}".format(i_episode, num_episodes), end="")
            sys.stdout.flush()
            
    # plot performance
    plt.plot(np.linspace(0,num_episodes,len(avg_scores),endpoint=False), np.asarray(avg_scores))
    plt.xlabel('Episode Number')
    plt.ylim([-250,0])
    plt.ylabel('Average Reward (Over Next %d Episodes)' % plot_freq)
    plt.show()
    # print best 100-episode performance
    print(('Best Average Reward over %d Episodes: ' % plot_freq), np.max(avg_scores))
    
    return Q
````
